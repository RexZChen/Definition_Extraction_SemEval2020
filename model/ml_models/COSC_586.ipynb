```python
from nltk.tokenize import word_tokenize
```

```python
'''文件读取'''
data = []
f = open("train.txt", encoding='utf-8')
# print(f)
for line in f:
    data.append(line)
f.close()
print(data[0])
```

```python
'''去掉开头序号-> [[句子(单词间以,为分割)], 0/1]'''
training_set = []
for unit in data:
    temp = unit.split(' ')
    if temp[1].isdigit():
        temp = temp[3:-1]
        # training_set.append(([temp, unit[-3:-2]]))
        training_set.append([' '.join(temp), unit[-3:-2]])
    else:
        temp = temp[1:-1]
        # training_set.append(([temp, unit[-3:-2]]))
        training_set.append([' '.join(temp), unit[-3:-2]])


print(len(training_set))
```

```python
import nltk
nltk.download('punkt')
```

```python
'''test'''
tokenized_word = word_tokenize(training_set[0][0])
print(tokenized_word)
```

```python
'''for all training set:'''
tokenzied_training_set = []
for samples in training_set:
  temp = word_tokenize(samples[0])
  tokenzied_training_set.append([temp, samples[1]])

print(tokenzied_training_set[0])
```

```python
from nltk.probability import FreqDist
```

```python
'''test'''
fdist_demo = []

fdist_demo = FreqDist(tokenzied_training_set[0][0])
print(fdist_demo)
```

```python
'''for all tokenized_training_set:'''
fdist_set = []
for items in tokenzied_training_set:
  fdist = FreqDist(items[0])
  fdist_set.append(fdist)

for fd in fdist_set:
  print(fd.most_common(2))
```

```python
'''for all sentences'''
filtered_set_temp = []
filtered_training_set = []
for sets in tokenzied_training_set:
  for w in sets[0]:
    # print(w)
    if w not in stop_words:
      filtered_set_temp.append(w)
  filtered_training_set.append([filtered_set_temp, sets[1]])
  filtered_set_temp = []

print(tokenzied_training_set[0])
print(filtered_training_set[0])
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
'''stemming'''
ps = PorterStemmer()

stemmed_words_temp = []
stemmed_words_training_set = []
for units in filtered_training_set:
  for w in units[0]:
    stemmed_words_temp.append(ps.stem(w))
  stemmed_words_training_set.append([stemmed_words_temp, units[1]])
  stemmed_words_temp = []

print(filtered_training_set[0])
print(stemmed_words_training_set[0])
```
